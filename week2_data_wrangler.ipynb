# Aggregation: Summary Statistics by Shift and Zone

```python
# Load the dataset
df = pd.read_csv(data_path)

# Create summary statistics for all numeric columns
summary_table = (
    df.groupby(['Shift', 'Zone'])
      .agg(['mean', 'max', 'min'])
)

print(summary_table)
```
